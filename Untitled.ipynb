## Since the 'emp_title' is categorical column so let's try with Label Encoder
#### Ordinal encoding should be used for ordinal variables (where order matters, like cold , warm , hot ); vs Label encoding should be used for non-ordinal (aka nominal) variables (where order doesn't matter, like blonde , brunette )

## For input features use Label Encoder and for target use Ordinal Encoder

```python
from sklearn.preprocessing import LabelEncoder

for col in ["emp_title"]:
    df[col] = LabelEncoder().fit_transform(df[col])
```

```python
sns.distplot(df['emp_title'])
```

```python
df['emp_title'].isnull().sum()
```

```python
df['emp_title']
```

```python
## from above we can see that 28027 is the value assigned to the null value
```

```python
df['emp_title'] = df['emp_title'].replace(to_replace=28027, value=np.nan)
```

```python
df['emp_title']
```

#### Distribution before handling null values

```python
sns.distplot(df['emp_title'])
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
imp_mean = imp_mean.fit(df[['emp_title']])
```

```python
emp_title_mean = imp_mean.transform(df[['emp_title']])
```

```python
emp_title_mean
```

```python
sns.distplot(emp_title_mean)
```

#### There is a change in distribution through mean imputation

```python
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
```

```python
imp_median = imp_median.fit(df[['emp_title']])
```

```python
emp_title_median = imp_median.transform(df[['emp_title']])
```

```python
emp_title_median
```

```python
sns.distplot(emp_title_median)
```

#### There is a change in distribution through median imputation (similar to mean imputation)

```python
imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mode = imp_mode.fit(df[['emp_title']])
emp_title_mode = imp_mode.transform(df[['emp_title']])
```

```python
emp_title_mode
```

```python
sns.distplot(emp_title_mode)
```

#### Better than mean and median. The distribution curve looks similar to the original one
